## Load Dataset

```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Basic info & missing values

```python
df.info()
df.isnull().sum()
```

# **Data Cleaning**

## Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

## Handle missing values

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)

df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', format='mixed')
df['date_added'].fillna(pd.Timestamp('1900-01-01'), inplace=True)

df.isnull().sum()
```

# **Exploratory Data Analysis (EDA)**

## Movies vs TV Shows

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar', color=['red', 'blue'])
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

## Top 10 countries

```python
top_countries = df[df['country'] != "Unknown"]['country'].value_counts().head(10).plot(kind='barh', color='green')
plt.title("Top 10 Countries on Netflix")
plt.xlabel("Number of Titles")
plt.show()
```

## Content added per year

```python
content_per_year = df['release_year'].value_counts().sort_index()
content_per_year = content_per_year[content_per_year.index >= 1980]

plt.figure(figsize=(12,6))
plt.plot(content_per_year.index, content_per_year.values, marker='o', linestyle='-', color='royalblue', linewidth=2)

peak_year = content_per_year.idxmax()
peak_value = content_per_year.max()
plt.scatter(peak_year, peak_value, color='red', s=100, label=f'Peak: {peak_year} ({peak_value})')

plt.title("Netflix Content Releases by Year (19802020)", fontsize=16, weight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Titles", fontsize=14)

plt.grid(alpha=0.3)
plt.legend()

plt.show()
```

## Most popular genres

```python
import seaborn as sns

genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)

sns.barplot(x=genres.values, y=genres.index, palette="viridis")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```
